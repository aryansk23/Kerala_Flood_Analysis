```python
monthly_avg = df[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].mean()
plt.figure(figsize=(12,6))
sns.barplot (x = monthly_avg.index, y =monthly_avg.values, color = 'cyan')
plt.title('Average monthly rainfall in Kerala (1901 - 2018)' , fontsize = 16)
plt.xlabel('Month')
plt.ylabel('Average rainfall (mm)')
plt.show()
```

### July has the highest average rainfall from the year 1901 to 2018

```python
plt.figure(figsize=(12,6))
sns.lineplot(data = df, x = 'YEAR', y = 'JUN', color = 'orchid', marker = 'D')
plt.title('July rainfall trend over the years in Kerala', fontsize = 16)
plt.xlabel('Year')
plt.ylabel('Rainfall in June(mm)')
plt.grid(True)
plt.show()
```

### The graph shows the up's and down's of the rainfall in July

```python
df['Decade'] = (df['YEAR'] // 10) * 10
plt.figure(figsize=(11,6))
custom_palette = ['slateblue', 'grey']
sns.countplot(x ='Decade',hue ='FLOODS', data = df, palette = custom_palette)
plt.title('Flood occurrences by decade', fontsize = 16)
plt.xlabel('Decade')
plt.ylabel('Count')
plt.legend(title = 'Floods')
plt.show()
```

### 1980 to 1990 had the least flood year

```python
df['Monsoon_Rainfall'] = df[['JUN', 'JUL', 'AUG', 'SEP']].sum(axis = 1)
df['Non_Monsoon_Rainfall'] = df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'OCT', 'NOV', 'DEC']].sum(axis = 1)
```

```python
plt.figure(figsize = (30,15))
df[['YEAR', 'Monsoon_Rainfall', 'Non_Monsoon_Rainfall']].set_index('YEAR').plot(kind = 'line', color=['slateblue', 'darkkhaki'])
plt.title('Monsoon vs Non-Monsoon Rainfall in Kerala', fontsize = 16)
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()
```
